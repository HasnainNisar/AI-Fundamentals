```python
import numpy as np

# 1. Creating Arrays
a = np.array([1, 2, 3, 4, 5])
print("1D Array:", a)

b = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", b)

# 2. Array Properties
print("Shape:", b.shape)
print("Size:", b.size)
print("Data Type:", b.dtype)

# 3. Array Operations
print("Sum:", np.sum(a))
print("Mean:", np.mean(a))
print("Max:", np.max(a))
print("Min:", np.min(a))

# 4. Zeros and Ones
zeros = np.zeros((3, 3))
ones = np.ones((2, 4))
print("Zeros:\n", zeros)
print("Ones:\n", ones)

# 5. Random Array
rand = np.random.rand(3, 3)
print("Random Array:\n", rand)

# 6. Reshaping
c = np.arange(12).reshape(3, 4)
print("Reshaped Array:\n", c)

# 7. Slicing
print("First row:", c[0])
print("Element at [1,2]:", c[1, 2])
```

```python
# 8. Matrix Operations
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print("Matrix Addition:\n", a + b)
print("Matrix Multiplication:\n", np.dot(a, b))
print("Transpose:\n", a.T)

# 9. Boolean Indexing
arr = np.array([10, 20, 30, 40, 50])
print("Elements greater than 25:", arr[arr > 25])

# 10. Statistical Operations
data = np.array([4, 7, 13, 2, 1, 8, 9, 3])
print("Sorted:", np.sort(data))
print("Standard Deviation:", np.std(data))
print("Variance:", np.var(data))
```
